```python
#def a functioninorder to 
def ALCOHOLDIST (row):
    if row['ALC4MOOD'] ==1 :
        return 5
    if row['ALCL_12MO'] == 1 :
        return 1
    if row['ALCL_12MO'] == 2 :
        return 2
    if row['ALCL_12MO'] == 3 :
        return 3
    if row['ALCL_12MO'] == 4 :
        return 4
```
